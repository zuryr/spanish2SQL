```python
import json
import time
from googletrans import Translator
```

```python
path = "../data/external/spider/tables.json"

# Abre el archivo JSON en modo lectura
with open(path, "r") as archivo:
    # Carga el contenido del archivo JSON en un diccionario
    datos = json.load(archivo)
```

```python
for i in datos:
    print(i["table_names"])
```

```python
j = 0
translator = Translator()
for i in datos:
    translated_tables_names = []
    for m in i["table_names"]:
        text_to_translate = m

        translated_text = translator.translate(text_to_translate, src='en', dest='es')

        translated_tables_names.append(translated_text.text)
        time.sleep(1)

    j += 1
    print(j)
    i["table_names"] = translated_tables_names
    
```

```python
for i in datos:
    print(i["table_names"])
```

```python
path_to_save = "../data/processed/"
file_name = "tables_spanish.json"

# Abrir el archivo en modo escritura
with open(path_to_save + file_name, "w",encoding="utf-8") as archivo:
    # Utilizar json.dump() para escribir el diccionario en el archivo
    json.dump(datos, archivo, indent=2, ensure_ascii=False)
```
